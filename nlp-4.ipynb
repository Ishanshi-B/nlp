```python
import string
def clean_punctuation(text):
    cleaned = "".join([i for i in text if i not in string.punctuation])
    return cleaned
datafile['textCleaned'] = datafile['text'].apply(lambda x:clean_punctuation(x))
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

def removeStopwords(text):
    words = [w for w in text if w not in stopwords.words('english')]
    return words

datafile['textRemovestopwords'] = datafile['textTokenized'].apply(lambda x: removeStopwords(x))

```

```python
#Lemmatize the data
from nltk.stem import WordNetLemmatizer
# Or use the stemmer instead
import nltk 
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem.porter import PorterStemmer

lemmatizer = WordNetLemmatizer()

def textLemmatizer(text):
    textLemmatized = [lemmatizer.lemmatize(i) for i in text]
    return textLemmatized

datafile['textLemmatized'] = datafile['textRemovestopwords'].apply(lambda x: textLemmatizer(x))
```

```python
def removeVerbs(text_file):

    # Tokenize the text
    tokens = nltk.word_tokenize(text_file)

    # Remove the verbs
    verbed = ['VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
    noVerbs = [word for word, tag in nltk.pos_tag(tokens) if tag not in verbed]

    # Join the tokens
    finishedNoVerbs = ' '.join(noVerbs)
    return finishedNoVerbs
```

```python
import re
def removeNumbers(text):

    # Remove the numbers
    noNums = re.sub('\d+', '', text)

    return noNums
```

```python
def removeNonWords(text): 
    worded = set(nltk.corpus.words.words())
    cleaning = "".join(w for w in nltk.wordpunct_tokenize(text) if w.lower() in worded or not w.isalpha())
    return cleaning
```

```python
def remove_non_real_words(text):
    # Tokenize the text
    tokens = nltk.word_tokenize(text)

    # Remove the non real words
    real_words = set(nltk.corpus.words.words())
    filtered_tokens = [word for word in tokens if word.lower() not in real_words]

    # Join the tokens
    filtered_text = ' '.join(filtered_tokens)

    return filtered_text
```
